```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
model = StandardScaler()
normalized_data = model.fit_transform(df_market_data)
df_normalized = pd.DataFrame(normalized_data, columns=df_market_data.columns, index=df_market_data.index)
df_normalized.head(10)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(df_normalized.values, columns=df_normalized.columns)

# Copy the crypto names from the original data

df_scaled['coid_id'] = df_market_data.index

# Set the coinid column as index

df_scaled.set_index('coid_id', inplace=True)

# Display sample data
df_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,11))
```

```python
# Create an empty list to store the inertia values

inertia_values = []
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:

for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(df_scaled) 
    inertia_values.append(model.inertia_) 

# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data) 
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve') 
plt.show()

# the different values of k to visually identify the optimal value for k.
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca = PCA(n_components=3)
# View the first five rows of the DataFrame. 
pca_data = pca.fit_transform(df_scaled)

df_pca = pd.DataFrame(pca_data, columns=['PC1','PC2','PC3' ])
df_pca.head(5)
```

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

for k in range(1,11): 
    model = KMeans(n_clusters=k) 
    model.fit(df_pca) 
    inertia_values.append(model.inertia_) 

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**
  * The point on the Elbow curve where the inertia (or within-cluster sum of squares) begins to decrease more gradually is known as the optimal k. In this instance, the Elbow curve indicates that k is best set to around 4, where the curve begins to flatten.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**
  We can determine whether the best value for k varies between the PCA data and the original data by comparing their Elbow curves and analyzing the corresponding k values. If the optimal k values from the two curves are different, it indicates that the number of ideal clusters differs between the original and PCA-transformed data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k = 4
kmeans_model = KMeans(n_clusters=k, random_state=42)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters = kmeans_model.predict(df_pca) 
print(clusters)
# Print the resulting array of cluster values.
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

plt.scatter(df_pca_copy['PC1'], df_pca_copy['PC2'], c=df_pca_copy['cluster'], cmap='Set1') 

plt.title("PCA Scatter Plot") 
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:**
Reduced Dimensionality: Utilizing fewer features leads to a reduction in data dimensionality. This simplification can enhance the clustering process by emphasizing the most crucial aspects of the data while removing noise and irrelevant information.

Loss of Information: Reducing the number of features might result in the loss of valuable information for clustering. If the excluded features contain significant patterns or contribute substantially to the data's overall structure, their omission may lead to less accurate or distinct clusters.

Increased Interpretability: With fewer features, the resulting clusters become more interpretable and easier to understand. High-dimensional data can be challenging to visualize and interpret, but reducing dimensionality simplifies the analysis and explanation of clustering results.

Improved Scalability: Clustering algorithms, like K-Means, can be computationally demanding for high-dimensional datasets. By decreasing the number of features, the scalability of the algorithm may be improved, enabling it to efficiently handle larger datasets.

Greater Sensitivity to Feature Selection: The selection of features becomes more critical when using fewer features. It's essential to choose features that have strong discriminatory power and effectively capture the data's core characteristics. Poor feature selection can negatively affect clustering outcomes.

In summary, using fewer features for clustering with K-Means offers both benefits and drawbacks. It can provide simplification, interpretability, and scalability advantages while posing risks of information loss and increased sensitivity to feature selection. The impact depends on the dataset, chosen features, and clustering objectives.
